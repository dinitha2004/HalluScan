# 2 · Honest out-of-fold evaluation + fusion

The re-trained heads in notebook 1 were fit on **all** rows, so scoring those same rows is **in-sample** — the 135168-dim SEP probe memorises its training set (AUROC 1.0 = leakage, not skill). This notebook evaluates **out-of-fold** instead:

1. reproduce TSV's exact 25% held-out split (so TSV's margins on the test rows are already OOF),
2. re-fit SEP + HalluShift on the **train** rows only and score the **test** rows,
3. train the fusion on train (with an inner 5-fold OOF so it isn't fed a memorised SEP feature),
4. report honest AUROC / AUPR / F1 on the held-out test rows.
